```python
import matplotlib.pyplot as plt
import numpy as np

# Linestyle
# use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:

# Use a dotted line:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.savefig('./image/sixteenth.jpg')
plt.show()
```

```python
# another style
# Use a dashed line:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed')
plt.savefig('./image/seventeenth.jpg')
plt.show()
```

```python
# Shorter Syntax
# The line style can be written in a shorter syntax:
# linestyle can be written as ls
# dotted can be written as :
# dashed can be written as --

# Shorter syntax: dotted

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = ':')
plt.savefig('./image/eighteenth.jpg')
plt.show()
```

```python
# Line Styles
# Style                   Or
# 'solid' (default)       '-'
# 'dotted'                ':'
# 'dashed'                '--'
# 'dashdot'               '-.'
# 'None'                  '' or ' '
```

```python
# Line Color
# use the keyword argument color or the shorter c to set the color of the line:

# Set the line color to red:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.savefig('./image/ninteenth.jpg')
plt.show()
```

```python
# also use Hexadecimal color values:
# Plot with a beautiful green line:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, c = '#4CAF50')
plt.savefig('./image/twentieth.jpg')
plt.show()
```

```python
# any of the 140 supported color names
# Plot with the color named "hotpink":

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, c = 'hotpink')
plt.savefig('./image/twentyoneth.jpg')
plt.show()
```

```python
# Line Width
# use the keyword argument linewidth or the shorter lw to change the width of the line.
# The value is a floating number, in points:

# Plot with a 20.5pt wide line:

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.savefig('./image/twentytowth.jpg')
plt.show()
```

```python
# Multiple Lines
# plot as many lines as you like by simply adding more plt.plot() functions:

# Draw two lines by specifying a plt.plot() function for each line:

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.savefig('./image/twentythreeth.jpg')
plt.show()
```

```python
# also plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.
# (In the examples above only specified the points on the y-axis, meaning that the points on the x-axis got the the default values (0, 1, 2, 3).)
# but the x- and y- values come in pairs:

# Draw two lines by specifiyng the x- and y-point values for both lines:

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.savefig('./image/twentyfourth.jpg')
plt.show()
```
